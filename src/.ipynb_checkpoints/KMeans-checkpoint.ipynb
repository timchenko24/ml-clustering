```python
def label_points(df, x, y, i, plt, legend):
    for point in range(0, df.shape[0]):
        plt.text(df[x][point], df[y][point], legend.loc[df[i][point]].id,
                 horizontalalignment='center', size='large', color='black')
```

# Experiment 1: Vessel Length - Voyage Profit

```python
length = length_profit['column_names'][0]
profit = length_profit['column_names'][1]

typ = length_profit['column_names'][2]
flag = length_profit['column_names'][3]

df = get_dataframe(conn_params, length_profit['query'], length_profit['column_names'])
```

```python
df.head()
```

**Sliced Dataframe for clusterization**

```python
df_c = df.loc[:,'length':'profit']
```

```python
df_c.head()
```

```python
type_legend = pd.DataFrame(index=df[typ].unique(), data=range(0, len(df[typ].unique())), columns=['id'])
```

## Original values

* ### WSS

```python
wss_list = []
```

```python
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++')
    kmeans.fit(df_c)
    wss_list.append(kmeans.inertia_)

plt.plot(range(1, 11), wss_list)
plt.xticks(range(1, 11))
plt.xlabel('Число кластеров(k)')
plt.ylabel('WSS')
plt.show()
```

* ### Sklearn K-Means (k = 3)

```python
kmeans = KMeans(n_clusters=3).fit(df_c)
centroids = kmeans.cluster_centers_
```

```python
sns.scatterplot(x=df_c[length], y=df_c[profit], hue=kmeans.labels_, palette='bright', s=50)
sns.scatterplot(x=centroids[:,0], y=centroids[:,1], marker='x', s=100, color='black', linewidth=1.2)
plt.xlabel('Длина судна')
plt.ylabel('Прибыль')
plt.legend(loc='upper left', frameon=True)
plt.show()
```

* ### Custom K-Means (k = 3)

```python
clusters, centers = custom_KMeans(k=3, dataset=df_c)
```

```python
visualize(3, clusters, centers, 'Длина судна', 'Прибыль')
```

* ### Vessel types distribution by clusters

```python
ax = sns.scatterplot(x=df_c[length], y=df_c[profit], hue=kmeans.labels_, palette='bright', s=50)

label_points(df, length, profit, typ, ax, type_legend)

sns.scatterplot(x=centroids[:,0], y=centroids[:,1], marker='x', s=100, color='red', linewidth=1.2)

plt.table(cellText=type_legend.values, rowLabels=type_legend.index, colLabels=type_legend.columns,
          cellLoc = 'center', rowLoc = 'center', loc='right', bbox=[1.4,0,.3,.9])
plt.xlabel('Длина судна')
plt.ylabel('Прибыль')
plt.legend(loc='upper left', frameon=True)
plt.show()
```

```python
type_cluster = pd.DataFrame(data={'type': df[typ], 'cluster': kmeans.labels_})
```

```python
type_cluster.groupby(['type', 'cluster']).size().reset_index(name='counts')
```

* ## Silhouette Method

```python
sil = []
kmax = 10

for k in range(2, kmax + 1):
    kmeans = KMeans(n_clusters = k).fit(df_c)
    labels = kmeans.labels_
    sil.append(silhouette_score(df_c, labels, metric = 'euclidean'))

plt.plot(range(2, kmax + 1), sil)
plt.xlabel('Число кластеров(k)')
plt.ylabel('Silhouette Score')
plt.show()
```

* ### Sklearn K-Means (k = 4)

```python
kmeans = KMeans(n_clusters=4).fit(df_c)
centroids = kmeans.cluster_centers_
```

```python
sns.scatterplot(x=df_c[length], y=df_c[profit], hue=kmeans.labels_, palette='bright', s=50)
sns.scatterplot(x=centroids[:,0], y=centroids[:,1], marker='x', s=100, color='black', linewidth=1.2)
plt.xlabel('Длина судна')
plt.ylabel('Прибыль')
plt.legend(loc='upper left', frameon=True)
plt.show()
```

```python
clusters, centers = custom_KMeans(k=4, dataset=df_c)
```

```python
visualize(4, clusters, centers, 'Длина судна', 'Прибыль')
```

```python
ax = sns.scatterplot(x=df_c[length], y=df_c[profit], hue=kmeans.labels_, palette='bright', s=50)

label_points(df, length, profit, typ, ax, type_legend)

sns.scatterplot(x=centroids[:,0], y=centroids[:,1], marker='x', s=100, color='red', linewidth=1.2)

plt.table(cellText=type_legend.values, rowLabels=type_legend.index, colLabels=type_legend.columns,
          cellLoc = 'center', rowLoc = 'center', loc='right', bbox=[1.4,0,.3,.9])
plt.xlabel('Длина судна')
plt.ylabel('Прибыль')
plt.legend(loc='upper left', frameon=True)
plt.show()
```
